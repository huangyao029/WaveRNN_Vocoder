```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(t, u, r):
    return 1 / (1 + np.exp(-(t - u) / r))


def sech(x):
    return 2 / (np.exp(x) + np.exp(-x))


def fff(x, u, s):
    return (1 / 4 * s) * sech((x - u) / (2 * s)) ** 2

x = np.arange(-10, 10, 0.1)
#y = logistic(x, 0, 3.3)
y = fff(x, 0, 3.3)

plt.figure(figsize = (10, 10))
plt.plot(x, y)
plt.show()
```
